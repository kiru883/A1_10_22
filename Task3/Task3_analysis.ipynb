# Solution

```python
PROJ_PATH = "/home/kirill/projects/personal_projects/A1_10_22/"
DATA_PATH = "/home/kirill/data/A1/"

tariff_plans_change_path = DATA_PATH + 'Tariff_plans_change.csv'
charges_path = DATA_PATH + 'Charges.csv'
suspended_path = DATA_PATH + 'Suspended.csv'

tariff_plans_change = pd.read_csv(tariff_plans_change_path)
charges = pd.read_csv(charges_path)
suspended = pd.read_csv(suspended_path)
```

### Directions of tariff's changes

```python
r = """
WITH bitariffs AS 
(
    SELECT DISTINCT SUBSCRIBER_ID, TARIFF_PLAN_ID, START_DTTM, END_DTTM
    FROM tariff_plans_change
    WHERE SUBSCRIBER_ID in 
    (
        SELECT SUBSCRIBER_ID
        FROM tariff_plans_change
        GROUP BY SUBSCRIBER_ID
        HAVING COUNT(SUBSCRIBER_ID) > 1
    )
),
tariff_changes AS
(
    SELECT SUBSCRIBER_ID, TARIFF_PLAN_ID, START_DTTM, END_DTTM,
    LEAD(TARIFF_PLAN_ID) OVER (PARTITION BY SUBSCRIBER_ID ORDER BY START_DTTM) as TARIFF_NEXT
    FROM bitariffs
),
tariff_filter AS
(
    SELECT SUBSCRIBER_ID, TARIFF_PLAN_ID, START_DTTM, END_DTTM, TARIFF_NEXT, TARIFF_PLAN_ID || ' -> ' || TARIFF_NEXT as TARIFF_CHANGE
    FROM tariff_changes
    WHERE TARIFF_NEXT IS NOT NULL AND TARIFF_PLAN_ID != TARIFF_NEXT
)



SELECT TARIFF_CHANGE, COUNT(*) as CHANGES_NUMBER
FROM tariff_filter
GROUP BY TARIFF_CHANGE
ORDER BY CHANGES_NUMBER DESC
"""

changes_df = ps(r)
print("Number of changes of tariffs\n\n", changes_df)

sns.barplot(changes_df, x='CHANGES_NUMBER', y='TARIFF_CHANGE');
```

### Account costs after tariff change

```python
r2 = """
WITH bitariffs AS 
(
    SELECT DISTINCT SUBSCRIBER_ID, TARIFF_PLAN_ID, START_DTTM, date(END_DTTM) as END_DTTM
    FROM tariff_plans_change
    WHERE SUBSCRIBER_ID in 
    (
        SELECT SUBSCRIBER_ID
        FROM tariff_plans_change
        GROUP BY SUBSCRIBER_ID
        HAVING COUNT(SUBSCRIBER_ID) > 1
    )
),
tariff_filter AS
(
    SELECT SUBSCRIBER_ID, 
    TARIFF_PLAN_ID || ' -> ' || TARIFF_NEXT as TARIFF_CHANGE, 
    date(strftime('%Y-%m-01',END_DTTM)) as END_MONTH,
    date(strftime('%Y-%m-01', END_DTTM), '-1 month') as PREV_MONTHS_KEY,
    date(strftime('%Y-%m-01', END_DTTM), '+1 month') as NEXT_MONTHS_KEY
    FROM 
    (
        SELECT SUBSCRIBER_ID, TARIFF_PLAN_ID, START_DTTM, END_DTTM,
        LEAD(TARIFF_PLAN_ID) OVER (PARTITION BY SUBSCRIBER_ID ORDER BY START_DTTM) as TARIFF_NEXT
        FROM bitariffs
    )
    WHERE TARIFF_NEXT IS NOT NULL AND TARIFF_PLAN_ID != TARIFF_NEXT
),
costs_average AS
(
    SELECT *, 
    AVG(CHARGES) OVER
    (
        PARTITION BY SUBSCRIBER_ID 
        ORDER BY BILL_MONTH
        ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
    ) as prev3_month_costs,
    AVG(CHARGES) OVER
    (
        PARTITION BY SUBSCRIBER_ID 
        ORDER BY BILL_MONTH
        ROWS BETWEEN CURRENT ROW AND 2 FOLLOWING
    ) as next3_month_costs
    FROM charges
),
prev AS 
(
    SELECT 
    tariff_filter.SUBSCRIBER_ID, 
    tariff_filter.TARIFF_CHANGE, 
    tariff_filter.END_MONTH, 
    tariff_filter.PREV_MONTHS_KEY, 
    tariff_filter.NEXT_MONTHS_KEY, 
    costs_average.BILL_MONTH, 
    costs_average.prev3_month_costs
    FROM tariff_filter
    INNER JOIN costs_average ON tariff_filter.PREV_MONTHS_KEY = costs_average.BILL_MONTH AND tariff_filter.SUBSCRIBER_ID = costs_average.SUBSCRIBER_ID 
)

SELECT prev.SUBSCRIBER_ID, prev.TARIFF_CHANGE, prev.prev3_month_costs, costs_average.next3_month_costs
FROM prev
INNER JOIN costs_average ON prev.NEXT_MONTHS_KEY = costs_average.BILL_MONTH AND prev.SUBSCRIBER_ID = costs_average.SUBSCRIBER_ID 
"""


r3 = """
SELECT 
AVG(prev3_month_costs) as avg_prev3_months_costs, 
AVG(next3_month_costs) as avg_next3_months_costs,
(AVG(next3_month_costs) - AVG(prev3_month_costs)) / AVG(prev3_month_costs) as costs_change_in_precents
FROM mean_costs_after_tariff_change_df
"""


r4 = """
SELECT 
TARIFF_CHANGE, 
AVG(prev3_month_costs) as avg_prev3_months_costs, 
AVG(next3_month_costs) as avg_next3_months_costs,
(AVG(next3_month_costs) - AVG(prev3_month_costs)) / AVG(prev3_month_costs) as costs_change_in_precents
FROM mean_costs_after_tariff_change_df
GROUP BY TARIFF_CHANGE
ORDER BY costs_change_in_precents DESC
"""


mean_costs_after_tariff_change_df = ps(r2)
mean_prev_next_costs = ps(r3)
mean_prev_next_costs_by_tariff_change = ps(r4)

_, axs = plt.subplots(nrows=2, figsize=(12, 12))

print("Average 3-month change of account cost for previous and next months of tariff change\n\n", mean_prev_next_costs)
l = mean_prev_next_costs.columns[:2].tolist()
val = mean_prev_next_costs[l].iloc[0].tolist()
p = mean_prev_next_costs['costs_change_in_precents'].iloc[0]
sns.barplot(x=l, y=val, ax=axs[0]).set_title(f"Overall change of account cost. Percent change: {round(p*100, 3)}%");

print("Average 3-month change of account cost for previous and next month for directions of tariff change\n\n", mean_prev_next_costs_by_tariff_change)
sns.barplot(mean_prev_next_costs_by_tariff_change, x='costs_change_in_precents', y='TARIFF_CHANGE', ax=axs[1]).set_title(f"Change of account cost for each tariff change directions");
```

### Suspends after tariff change

```python
r5 = """
WITH bitariffs AS 
(
    SELECT DISTINCT SUBSCRIBER_ID, TARIFF_PLAN_ID, START_DTTM, date(END_DTTM) as END_DTTM
    FROM tariff_plans_change
    WHERE SUBSCRIBER_ID in 
    (
        SELECT SUBSCRIBER_ID
        FROM tariff_plans_change
        GROUP BY SUBSCRIBER_ID
        HAVING COUNT(SUBSCRIBER_ID) > 1
    )
),
tariff_filter AS
(
    SELECT SUBSCRIBER_ID, 
    date(strftime('%Y-%m-01', END_DTTM)) as END_DTTM,
    TARIFF_PLAN_ID || ' -> ' || TARIFF_NEXT as TARIFF_CHANGE
    FROM 
    (
        SELECT SUBSCRIBER_ID, TARIFF_PLAN_ID, START_DTTM, END_DTTM,
        LEAD(TARIFF_PLAN_ID) OVER (PARTITION BY SUBSCRIBER_ID ORDER BY START_DTTM) as TARIFF_NEXT
        FROM bitariffs
    )
    WHERE TARIFF_NEXT IS NOT NULL AND TARIFF_PLAN_ID != TARIFF_NEXT
),
tariff_with_suspendeds AS
(
    SELECT 
        tariff_filter.SUBSCRIBER_ID, 
        tariff_filter.END_DTTM,
        tariff_filter.TARIFF_CHANGE,
        suspended.START_DT,
        CASE
            WHEN suspended.START_DT BETWEEN date(tariff_filter.END_DTTM, '+1 month') AND date(tariff_filter.END_DTTM, '+4 month') THEN 1.0 
            WHEN suspended.START_DT IS NULL THEN 0.0
            ELSE 0.0
        END as next3_month_sus,
        CASE
            WHEN suspended.START_DT BETWEEN date(tariff_filter.END_DTTM, '-3 month') AND tariff_filter.END_DTTM THEN 1.0 
            WHEN suspended.START_DT IS NULL THEN 0.0
            ELSE 0.0
        END as prev3_month_sus
    FROM tariff_filter
    LEFT JOIN suspended ON tariff_filter.SUBSCRIBER_ID = suspended.SUBSCRIBER_ID
    
)

SELECT SUBSCRIBER_ID, TARIFF_CHANGE, SUM(prev3_month_sus) / 3 as avg_prev3_month_suspended, SUM(next3_month_sus) / 3 as avg_next3_month_suspended
FROM tariff_with_suspendeds
GROUP BY SUBSCRIBER_ID, TARIFF_CHANGE
ORDER BY SUBSCRIBER_ID
"""


r6 = """
SELECT AVG(avg_prev3_month_suspended) as avg_prev3_month_suspended, AVG(avg_next3_month_suspended) as avg_next3_month_suspended,
(AVG(avg_next3_month_suspended) - AVG(avg_prev3_month_suspended)) / AVG(avg_prev3_month_suspended) as sus_change_in_percents

FROM mean_sus_after_tariff_change_df
"""


r7 = """
SELECT TARIFF_CHANGE, AVG(avg_prev3_month_suspended) as avg_prev3_month_suspended, AVG(avg_next3_month_suspended) as avg_next3_month_suspended,
CASE
    WHEN AVG(avg_prev3_month_suspended) = 0 AND AVG(avg_next3_month_suspended) = 0 THEN 0.0
    WHEN AVG(avg_prev3_month_suspended) = 0 AND NOT AVG(avg_next3_month_suspended) = 0 THEN 1.0
    ELSE (AVG(avg_next3_month_suspended) - AVG(avg_prev3_month_suspended)) / AVG(avg_prev3_month_suspended)
END as sus_change_in_percents,
COUNT(TARIFF_CHANGE) as changes_number
FROM mean_sus_after_tariff_change_df
GROUP BY TARIFF_CHANGE
ORDER BY sus_change_in_percents DESC
"""

mean_sus_after_tariff_change_df = ps(r5)
mean_prev_next_sus = ps(r6)
mean_prev_next_sus_by_tariff_change = ps(r7)

_, axs = plt.subplots(nrows=2, figsize=(12, 12))

print("Average 3-month change of cases of user suspendeds for previous and next months of tariff change\n\n", mean_prev_next_sus)
l = mean_prev_next_sus.columns[:2].tolist()
val = mean_prev_next_sus[l].iloc[0].tolist()
p = mean_prev_next_sus['sus_change_in_percents'].iloc[0]
sns.barplot(x=l, y=val, ax=axs[0]).set_title(f"Overall change of cases of user suspendeds. Percent change: {round(p*100, 3)}%");

print("Average 3-month change of cases of user suspendeds for previous and next month for directions of tariff change\n\n", mean_prev_next_sus_by_tariff_change)
sns.barplot(mean_prev_next_sus_by_tariff_change, x='sus_change_in_percents', y='TARIFF_CHANGE', ax=axs[1]).set_title(f"Change of suspends for each tariff change directions");
```
